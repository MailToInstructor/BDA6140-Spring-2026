```python
#step  1  --> X and y
X = df[["YearsExp"]]
y = df ["Salary"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Step 2 --> create a model object using LinearRegreesion

salary_exp_model = LinearRegression()
```

```python
#step 3 --> Fit For training
salary_exp_model.fit(X,y)
```

```python
#step 4 --Extract the intercept and slope

b0 = salary_exp_model.intercept_
b1 = salary_exp_model.coef_[0]

print(b0, b1)
```

```python
# Last step --> Find r and R^2

r_squared = salary_exp_model.score(X,y)

r = df["Salary"].corr(df['YearsExp'])

print(r_squared, r)
```

```python
# what is the salray x = 4 ( 4 years of experience )

salary_4 = salary_exp_model.predict ([[4]])
salary_4
```

```python
# what is the residual when x = 4 --> residual = y - y_hat

#what is the y_hat --> salary_exp_model.predict ([[4]])
#what is the actualy y when x = 4 --> you can take a look at the datafram df

```
