## Problem 4  (10 points)

Run the code cell below to train a decision tree, then predict the result for **a new student with `StudyHours = 7`, `Attendance = 75`, `PracticeScore = 75`**.

What is the predicted result?

A. Pass
B. Fail
C. 1
D. Error

````python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

data = pd.DataFrame({
    "StudyHours":    [2, 3, 3, 4, 5, 5, 6, 7, 8, 9,
                      5, 6, 7, 8, 9, 10, 6, 7, 8, 9],
    "Attendance":    [45, 50, 55, 58, 60, 65, 68, 70, 72, 75,
                      78, 80, 82, 85, 88, 90, 60, 65, 70, 80],
    "PracticeScore": [30, 35, 40, 45, 50, 55, 60, 65, 68, 70,
                      75, 78, 80, 85, 88, 92, 72, 73, 73, 74],
    "Result":        ["Fail","Fail","Fail","Fail","Fail",
                      "Fail","Fail","Fail","Fail","Fail",
                      "Pass","Pass","Pass","Pass","Pass","Pass",
                      "Fail","Fail","Pass","Pass"]
})

X = data[["StudyHours", "Attendance", "PracticeScore"]]
y = data["Result"]

tree_clf = DecisionTreeClassifier(max_depth=3, random_state=42)
tree_clf.fit(X, y)
````
