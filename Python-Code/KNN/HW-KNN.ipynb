## Part 1: Load and Explore the Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Create the dataset (36 students)
data = pd.DataFrame({
    'StudyHours': [8, 9, 7, 10, 8, 9, 10, 8, 9, 7, 10, 7,
                   6, 5, 7, 8, 6, 5, 6, 8, 9, 5, 7, 6,
                   3, 2, 4, 3, 5, 6, 4, 2, 3, 4, 5, 4],
    'SleepHours': [7, 7, 8, 6, 8, 6, 7, 6, 8, 7, 8, 6,
                   7, 8, 5, 9, 6, 7, 8, 5, 9, 6, 9, 5,
                   7, 8, 5, 9, 4, 10, 6, 7, 5, 10, 10, 4],
    'Grade': ['A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A',
              'B', 'B', 'B', 'B', 'B', 'B', 'B', 'B', 'B', 'B', 'B', 'B',
              'C', 'C', 'C', 'C', 'C', 'C', 'C', 'C', 'C', 'C', 'C', 'C']
})

data
```

```python
# Visualize the data
plt.figure(figsize=(8, 6))

colors = {'A': 'green', 'B': 'blue', 'C': 'red'}

for grade in ['A', 'B', 'C']:
    subset = data[data['Grade'] == grade]
    plt.scatter(subset['StudyHours'], subset['SleepHours'],
                c=colors[grade], label=f'Grade {grade}', s=100, edgecolors='black')

plt.xlabel('Study Hours', fontsize=12)
plt.ylabel('Sleep Hours', fontsize=12)
plt.title('Student Grades by Study Hours and Sleep Hours', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

---
## Part 2: Prepare the Data

```python
# Define X and y
X = data[['StudyHours', 'SleepHours']]
y = data['Grade']

print("Features (X):")
print(X.head())
print("\nTarget (y):")
print(y.head())
```

```python
# Split into training and testing sets (stratified to keep class balance)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

print(f"Training samples: {len(X_train)}")
print(f"Testing samples: {len(X_test)}")
print(f"\nClass distribution in test set:")
print(y_test.value_counts())
```

```python
# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Features scaled!")
print(f"\nBefore scaling (first training sample): {X_train.iloc[0].values}")
print(f"After scaling (first training sample): {X_train_scaled[0]}")
```

---
## Part 3: Train the KNN Model

```python
# Train KNN with k=3
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_scaled, y_train)

print("KNN model trained with k=3")
```

```python
# Predict on testing data
y_pred = knn.predict(X_test_scaled)

results = pd.DataFrame({
    'Actual': y_test.values,
    'Predicted': y_pred
})

results
```

---
## Part 4: Evaluate the Model

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```

```python
# Build confusion matrix
labels = ['A', 'B', 'C']
cm = confusion_matrix(y_test, y_pred, labels=labels)

cm_df = pd.DataFrame(
    cm,
    index=['Actual: A', 'Actual: B', 'Actual: C'],
    columns=['Predicted: A', 'Predicted: B', 'Predicted: C']
)

cm_df
```

```python
# Precision, Recall, and F1-score
# Note: Focus on precision and recall. You can ignore F1-score for now.
print(classification_report(y_test, y_pred))
```

---
## Part 5: How Does k Impact Accuracy?

```python
# Test different values of k
# k=1
knn_1 = KNeighborsClassifier(n_neighbors=1)
knn_1.fit(X_train_scaled, y_train)
acc_1 = accuracy_score(y_test, knn_1.predict(X_test_scaled))
print(f"k = 1:  Accuracy = {acc_1:.4f}")

# k=3
knn_3 = KNeighborsClassifier(n_neighbors=3)
knn_3.fit(X_train_scaled, y_train)
acc_3 = accuracy_score(y_test, knn_3.predict(X_test_scaled))
print(f"k = 3:  Accuracy = {acc_3:.4f}")

# k=5
knn_5 = KNeighborsClassifier(n_neighbors=5)
knn_5.fit(X_train_scaled, y_train)
acc_5 = accuracy_score(y_test, knn_5.predict(X_test_scaled))
print(f"k = 5:  Accuracy = {acc_5:.4f}")

# k=7
knn_7 = KNeighborsClassifier(n_neighbors=7)
knn_7.fit(X_train_scaled, y_train)
acc_7 = accuracy_score(y_test, knn_7.predict(X_test_scaled))
print(f"k = 7:  Accuracy = {acc_7:.4f}")

# k=9
knn_9 = KNeighborsClassifier(n_neighbors=9)
knn_9.fit(X_train_scaled, y_train)
acc_9 = accuracy_score(y_test, knn_9.predict(X_test_scaled))
print(f"k = 9:  Accuracy = {acc_9:.4f}")

# k=11
knn_11 = KNeighborsClassifier(n_neighbors=11)
knn_11.fit(X_train_scaled, y_train)
acc_11 = accuracy_score(y_test, knn_11.predict(X_test_scaled))
print(f"k = 11: Accuracy = {acc_11:.4f}")

# k=13
knn_13 = KNeighborsClassifier(n_neighbors=13)
knn_13.fit(X_train_scaled, y_train)
acc_13 = accuracy_score(y_test, knn_13.predict(X_test_scaled))
print(f"k = 13: Accuracy = {acc_13:.4f}")

# k=15
knn_15 = KNeighborsClassifier(n_neighbors=15)
knn_15.fit(X_train_scaled, y_train)
acc_15 = accuracy_score(y_test, knn_15.predict(X_test_scaled))
print(f"k = 15: Accuracy = {acc_15:.4f}")
```

```python
# Plot accuracy vs k
k_values = [1, 3, 5, 7, 9, 11, 13, 15]
accuracies = [acc_1, acc_3, acc_5, acc_7, acc_9, acc_11, acc_13, acc_15]

plt.figure(figsize=(8, 5))
plt.plot(k_values, accuracies, marker='o', linewidth=2, markersize=8, color='blue')
plt.xlabel('k (number of neighbors)', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.title('How k Affects Accuracy', fontsize=14)
plt.xticks(k_values)
plt.grid(True, alpha=0.3)
plt.show()
```
