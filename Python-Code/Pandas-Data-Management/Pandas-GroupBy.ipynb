## What is `groupby`?

`groupby` follows the **split → apply → combine** idea:

1. **Split** the DataFrame into groups using a column (example: `Sex`)
2. **Apply** a summary function to each group (example: `mean`, `sum`, `count`)
3. **Combine** the results into a summary table

**Typical syntax**
```python
df.groupby("GroupColumn")["TargetColumn"].mean()
```


## Count rows in each group: `size()`

**Question:** How many customers are in each `Sex` group?


````python
df.groupby("Sex").size()
````

✅ **Note**
- `.size()` counts **all rows** in each group.


## Group average (mean) of one column

**Question:** What is the average `Spending2021` for each `Sex`?


````python
df.groupby("Sex")["Spending2021"].mean()
````

````python
# Average income by College
df.groupby("College")["Income"].mean()
````

✅ **Note**
- `.mean()` ignores missing values (`NaN`) automatically.


## `count()` vs `size()` (very important)

These two can be **different**.

- `size()` counts **all rows** in each group.
- `count()` counts **non-missing values** in a specific column (within each group).

**Example:** compare `count()` vs `size()` for `Spending2021`.


````python
# counts ONLY non-missing Spending2021 values in each Sex group
df.groupby("Sex")["Spending2021"].count()
````

````python
df.groupby("Sex")["Spending2021"].size()
````

````python
# counts ALL rows in each Sex group
df.groupby("Sex").size()
````

✅ **Takeaway**
- Use `size()` when you want **number of rows** in each group.
- Use `count()` when you want **number of non-missing values** in a specific column.


## Multiple statistics with `.agg()`

**Question:** For each `Sex`, show `count`, `mean`, `min`, and `max` of `Spending2021`.


````python
df.groupby("Sex")["Spending2021"].agg(["count", "mean", "min", "max"])
````

### Sort the result using sort_values()

````python
df.groupby("Channel").size()

print(df.groupby("Channel").size().sort_values(ascending=False))

print("")

print(df.groupby("Channel").size().sort_values(ascending=True))

print("")


print(df.groupby("Channel").size().sort_values()) #by default --> asending = True
````
