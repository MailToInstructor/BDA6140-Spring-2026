## Replace missing values with `fillna()` (use the mean)

✅ **Key point:** fillna() replace missing values with the value provided to fillna(). `fillna()` does **NOT** change the original DataFrame (`df`) by default (case 0)
To change data (orginal dataframe), you must either:
- **assign the result back**, (case 1) or
- use **`inplace=True`** (case 2)

We will show **three cases**.


### Case 0) `fillna()` without assignment (original `df` does NOT change)

First compute the mean (Pandas ignores NaN automatically):


```python
mean_2021 = df["Spending2021"].mean()
mean_2021
```

```python
# Missing values before
df["Spending2021"].isna().sum()
```

```python
# Fill missing values (but do NOT assign back)
temp = df["Spending2021"].fillna(mean_2021)     #does not update/change the orginal dataframe, df
                                                #temp has a new dataframe where all the rows with missing valus removed!

# Missing values in the new result
temp.isna().sum()
```

```python
# Prove original df is unchanged
df["Spending2021"].isna().sum()
```

✅ **Key point (Case 0):** `fillna()` returns a new result, but **does not change `df`** unless you assign it back (or use `inplace=True`).

### Case 1) Assign the result back (the original DataFrame df changes)

To keep the original `df` safe, use a copy:


```python
df_case1 = df.copy()  #I do not want to change the orginal data frame df.
                      # I make the copy --> df_case1

# Missing values before
df_case1["Spending2021"].isna().sum()
```

```python
mean_2021_case1 = df_case1["Spending2021"].mean()

# Assign the result back (this changes df_case1)
df_case1["Spending2021"] = df_case1["Spending2021"].fillna(mean_2021_case1) # note that df_case1 (spending2021) has been updated

# Missing values after
df_case1["Spending2021"].isna().sum()
```

✅ **Key point (Case 1):** The DataFrame changes because you **assign the result back**.

### Case 2) Use `inplace=True` (the DataFrame changes)

Again, use a copy:


```python
df_case2 = df.copy()

# Missing values before
df_case2["Spending2021"].isna().sum()
```

```python
mean_2021_case2 = df_case2["Spending2021"].mean()

# inplace=True changes df_case2 directly
df_case2.fillna(mean_2021_case2, inplace = True) #inplace = True --> it update the orginal dataframe, df_case2

# Missing values after
df_case2["Spending2021"].isna().sum()
```

✅ **Key point (Case 2):** The DataFrame changes because `inplace=True` modifies it directly.

## Fill multiple columns at once (each column uses its own mean)

✅ Multiple columns must use **double square brackets**.


```python
df_fill_means = df.copy()

cols = ["Spending2020", "Spending2021"]

# Each column is filled using its own mean
df_fill_means[cols] = df_fill_means[cols].fillna(df_fill_means[cols].mean())

df_fill_means[cols].isna().sum()
```
