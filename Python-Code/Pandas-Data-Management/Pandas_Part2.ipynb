# Section 4. Row subsetting using boolean filters

## What is Row Filtering Using Boolean Expressions?

**Row filtering** (also called **boolean indexing**) is a way to select only the rows from a DataFrame that satisfy a given condition.

When you write a comparison like `df["Income"] >= 80000`, Pandas evaluates it for **every row** and returns a **Series of True/False values** (one per row). When you place that Series inside `df[...]`, Pandas keeps only the rows where the value is `True` and drops the rows where it is `False`.

**Typical syntax:**

```
df[ boolean_expression ]
```

How it works step-by-step:

| Step | Code | Result |
|------|------|--------|
| 1. Write a condition | `df["Income"] >= 80000` | A Series of `True` / `False` for every row |
| 2. Pass it into `df[...]` | `df[ df["Income"] >= 80000 ]` | A new DataFrame containing **only** the rows where the condition is `True` |

**Think of it as:** `df[ True/False for every row ]` → returns only the `True` rows.

````python
df.head()
````

````python
boolean_exp = df["Income"] >= 80000  #--> boolean expression/filter
boolean_exp
````

````python
df[ boolean_exp  ]
````

````python
df[ df["Income"] >= 80000  ]
````

## 4.2 Multiple conditions
- Use `&` for AND, `|` for OR, `~` for NOT
- Put each condition in parentheses

### Boolean Operator Reference Tables

Below are the truth tables for the three logical operators used in Pandas row filtering.

**AND (`&`)** — Both conditions must be `True`

| A | B | A & B |
|---|---|-------|
| True | True | **True** |
| True | False | False |
| False | True | False |
| False | False | False |

**OR (`|`)** — At least one condition must be `True`

| A | B | A \| B |
|---|---|--------|
| True | True | **True** |
| True | False | **True** |
| False | True | **True** |
| False | False | False |

**NOT (`~`)** — Flips `True` ↔ `False`

| A | ~A |
|---|-----|
| True | False |
| False | **True** |

## 4.3 Row-Column Subsetting with `iloc` and `loc`

So far we have learned:
- **Column subsetting:** `df["col"]` or `df[["col1", "col2"]]` → select columns
- **Row subsetting:** `df[boolean_expression]` → filter rows

But what if you want to select **specific rows AND specific columns at the same time**?

Pandas provides two indexers for this:

| Indexer | Selects by | Syntax |
|---------|-----------|--------|
| `df.iloc[row, col]` | **Integer position** (0, 1, 2, …) | `df.iloc[row_position, col_position]` |
| `df.loc[row, col]` | **Label / index name** | `df.loc[row_label, col_name]` |

**Think of it as:**
- **`iloc`** = **i**nteger **loc**ation (use numbers)
- **`loc`** = **l**abel-based **loc**ation (use index labels and column names)

### 4.3.1 `iloc` — Selection by integer position

**Q: What is the value in the 1st row and 1st column of the DataFrame (i.e., position [0, 0])?**

````python
df.head()
````

````python
# Single element: row 0, column 0
df.iloc[0, 0]
````

**Q: What are the first 5 rows and first 3 columns of the DataFrame?**

````python
# First 5 rows, first 3 columns (using slicing)
df.iloc[0:5, 0:3]
````

**Q: What are the values in rows 0, 5, and 10 for columns at positions 0, 7, and 13 (`CustID`, `Income`, and `Channel`)?**

````python
df.head()
````

````python
# Specific rows and columns by position (using lists)
df.iloc[[0, 5, 10], [0, 7, 13]]
````

**Q: What are all columns for the first 3 rows?**

*Tip:* Use `:` to mean "all" for either rows or columns.

````python
# All columns for the first 3 rows
df.iloc[0:3, :]
````

### 4.3.2 `loc` — Selection by label / column name

**Q: What are the `CustID`, `Income`, and `Channel` values for the row with index label 0?**

````python
# Single row by index label, specific columns by name
df.loc[0, ["CustID", "Income", "Channel"]]
````

**Q: What are the `Sex`, `Income`, and `Satisfaction` values for rows with index labels 0 through 4?**

*Note:* Unlike `iloc`, `loc` slicing is **inclusive** on both ends — `0:4` returns rows 0, 1, 2, 3, **and** 4.

````python
# Rows 0 to 4 (inclusive!), selected columns by name
df.loc[0:4, ["Sex","Income", "Satisfaction"]]
````

````python
#please use idential expression using iloc (good exam question)
df.iloc[0:5    , [1,7, 12]  ]
````

**Q: What are the `CustID` and `Income` values for all rows (using `loc`)?**

````python
# All rows, selected columns by name
df.loc[:, ["CustID", "Income"]].head()
````

### 4.3.3 Combining boolean filters with `loc`

You can also use a **boolean expression** as the row selector inside `loc` to filter rows AND select specific columns at the same time.

```
df.loc[boolean_expression, [col1, col2, ...]]
```

**Q: For customers whose `Income` is greater than or equal to 100,000, what are their `CustID`, `Income`, and `Channel` values?**

````python
# Boolean filter + column selection using loc
df.loc[df["Income"] >= 100000, ["CustID", "Income", "Channel"]].head()
````

**Q: For customers who are `Female` and have `Spending2021` greater than 800, what are their `CustID`, `Sex`, `Spending2021`, and `Satisfaction` values?**

````python
# Multiple boolean conditions + column selection
df.loc[(df["Sex"] == "Female") & (df["Spending2021"] > 800),  ["CustID", "Sex", "Spending2021", "Satisfaction"]]
````

### `iloc` vs `loc` — Quick Comparison

| Feature | `iloc` | `loc` |
|---------|--------|-------|
| Selects by | Integer position (0, 1, 2, …) | Label / column name |
| Slicing end | **Exclusive** (`0:3` → rows 0, 1, 2) | **Inclusive** (`0:3` → rows 0, 1, 2, 3) |
| Column reference | Column number (e.g., `7`) | Column name (e.g., `"Income"`) |
| Boolean filter as row selector | ❌ Not supported | ✅ Supported |
