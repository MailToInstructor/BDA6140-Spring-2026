## Optional Exploration — Not Graded

The cell below plots the **fitted logistic regression curve** for the tumor dataset,
together with two decision thresholds: **0.5** (default) and **0.3** (lowered).

- The **gray dashed line** at $p = 0.5$ is the default threshold.
  Tumors to the right of its vertical marker (~3.46 cm) are predicted **malignant**.
- The **orange dashed line** at $p = 0.3$ is the lowered threshold.
  Tumors to the right of its vertical marker (~2.78 cm) are now predicted **malignant**.
- The **shaded zone** between the two vertical markers shows the tumors
  whose prediction **changes** from benign to malignant when the threshold is lowered.

Run this cell after you have trained `model` in the cell above Question 11.

````python
import math
import numpy as np
import matplotlib.pyplot as plt

# Fitted curve on a dense x-grid
x_range = np.linspace(0.5, 9.5, 300).reshape(-1, 1)
p_range = model.predict_proba(x_range)[:, 1]

# Decision boundary for each threshold
b0 = model.intercept_[0]
b1 = model.coef_[0][0]
boundary_05 = -b0 / b1
boundary_03 = (math.log(0.3 / 0.7) - b0) / b1

fig, ax = plt.subplots(figsize=(10, 5))

# Data points
ax.scatter(tumor_size[diagnosis == 0], diagnosis[diagnosis == 0],
           color='blue', s=40, label='Benign (0)', zorder=5)
ax.scatter(tumor_size[diagnosis == 1], diagnosis[diagnosis == 1],
           color='red',  s=40, label='Malignant (1)', zorder=5)

# Logistic curve
ax.plot(x_range, p_range, color='green', linewidth=2, label='P(Malignant)')

# Horizontal threshold lines
ax.axhline(y=0.5, linestyle='--', color='gray',   linewidth=1.5,
           label=f'Threshold = 0.5  (boundary ≈ {boundary_05:.2f} cm)')
ax.axhline(y=0.3, linestyle='--', color='orange', linewidth=1.5,
           label=f'Threshold = 0.3  (boundary ≈ {boundary_03:.2f} cm)')

# Vertical boundary lines
ax.axvline(x=boundary_05, linestyle=':', color='gray',   linewidth=1.2)
ax.axvline(x=boundary_03, linestyle=':', color='orange', linewidth=1.2)

# Shaded zone — predictions change here
ax.axvspan(boundary_03, boundary_05, alpha=0.12, color='orange',
           label='Zone where prediction changes\n(benign → malignant)')

# Annotations
ax.annotate(f'Boundary\n≈ {boundary_05:.2f} cm',
            xy=(boundary_05, 0.5), xytext=(boundary_05 + 0.4, 0.65),
            arrowprops=dict(arrowstyle='->', color='gray'),
            color='dimgray', fontsize=9)
ax.annotate(f'Boundary shifts\nto ≈ {boundary_03:.2f} cm',
            xy=(boundary_03, 0.3), xytext=(boundary_03 - 2.2, 0.45),
            arrowprops=dict(arrowstyle='->', color='orange'),
            color='darkorange', fontsize=9)

ax.set_xlabel('Tumor Size (cm)')
ax.set_ylabel('Predicted Probability of Malignant')
ax.set_title('Effect of Lowering the Decision Threshold from 0.5 to 0.3')
ax.set_ylim(-0.05, 1.10)
ax.legend(loc='upper left', fontsize=9)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f'Threshold 0.5  →  decision boundary at {boundary_05:.2f} cm')
print(f'Threshold 0.3  →  decision boundary shifts to {boundary_03:.2f} cm')
print(f'Tumors between {boundary_03:.2f} cm and {boundary_05:.2f} cm',
      'switch from benign to malignant.')
````
